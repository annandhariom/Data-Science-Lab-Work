## 1. Variables and Data Types
Python supports several built-in data types. Variables are created simply by assigning a value — no declaration needed.

```python
# Integer
age = 21
print("Age:", age, "| Type:", type(age))

# Float
gpa = 3.85
print("GPA:", gpa, "| Type:", type(gpa))

# String
name = "Alice"
print("Name:", name, "| Type:", type(name))

# Boolean
is_student = True
print("Is Student:", is_student, "| Type:", type(is_student))

# NoneType
grade = None
print("Grade:", grade, "| Type:", type(grade))
```

## 2. Arithmetic Operators
Python supports standard arithmetic operations.

```python
a = 15
b = 4

print(f"a = {a}, b = {b}")
print(f"Addition:       a + b = {a + b}")
print(f"Subtraction:    a - b = {a - b}")
print(f"Multiplication: a * b = {a * b}")
print(f"Division:       a / b = {a / b}")
print(f"Floor Division: a // b = {a // b}")
print(f"Modulus:        a % b = {a % b}")
print(f"Exponentiation: a ** b = {a ** b}")
```

## 3. Comparison and Logical Operators

```python
x = 10
y = 20

print("Comparison Operators:")
print(f"x == y : {x == y}")
print(f"x != y : {x != y}")
print(f"x < y  : {x < y}")
print(f"x > y  : {x > y}")
print(f"x <= y : {x <= y}")
print(f"x >= y : {x >= y}")

print("\nLogical Operators:")
print(f"x < 15 and y > 15 : {x < 15 and y > 15}")
print(f"x > 15 or y > 15  : {x > 15 or y > 15}")
print(f"not (x == y)       : {not (x == y)}")
```

## 4. Strings and String Methods
Strings in Python are immutable sequences of characters. Python provides a rich set of built-in string methods.

```python
sentence = "  Hello, Python World!  "

print("Original       :", repr(sentence))
print("Stripped       :", sentence.strip())
print("Upper          :", sentence.strip().upper())
print("Lower          :", sentence.strip().lower())
print("Title Case     :", sentence.strip().title())
print("Replace        :", sentence.strip().replace("Python", "Data Science"))
print("Split          :", sentence.strip().split(","))
print("Starts with 'H':", sentence.strip().startswith("H"))
print("Contains 'World':", "World" in sentence)
print("Length         :", len(sentence.strip()))
print("Char at index 0:", sentence.strip()[0])
print("Slice [0:5]    :", sentence.strip()[0:5])
```

```python
# String Formatting
name = "Bob"
score = 92.5

# f-string (recommended)
print(f"Student {name} scored {score:.1f}%")

# .format() method
print("Student {} scored {:.1f}%".format(name, score))

# % operator (older style)
print("Student %s scored %.1f%%" % (name, score))
```

## 5. Lists
Lists are ordered, mutable sequences that can hold elements of different types.

```python
# Creating a list
fruits = ["apple", "banana", "cherry", "date", "elderberry"]
print("List:", fruits)
print("Length:", len(fruits))
print("First element:", fruits[0])
print("Last element:", fruits[-1])
print("Slice [1:3]:", fruits[1:3])

# Modifying
fruits.append("fig")
print("After append:", fruits)

fruits.insert(2, "blueberry")
print("After insert at index 2:", fruits)

fruits.remove("banana")
print("After remove 'banana':", fruits)

popped = fruits.pop()
print(f"Popped '{popped}', list now:", fruits)

fruits.sort()
print("Sorted:", fruits)

fruits.reverse()
print("Reversed:", fruits)
```

## 6. Tuples
Tuples are ordered and **immutable** — they cannot be changed after creation.

```python
coordinates = (40.7128, -74.0060)  # New York City lat/lon
print("Tuple:", coordinates)
print("Latitude:", coordinates[0])
print("Longitude:", coordinates[1])

# Unpacking
lat, lon = coordinates
print(f"Unpacked → Lat: {lat}, Lon: {lon}")

# Tuple with mixed types
student = ("Alice", 21, 3.85, True)
print("\nStudent record:", student)
print("Count of elements:", len(student))
print("Is 3.85 in tuple:", 3.85 in student)
```

## 7. Dictionaries
Dictionaries store data as key-value pairs and are unordered (insertion-ordered in Python 3.7+).

```python
student = {
    "name": "Alice",
    "age": 21,
    "major": "Data Science",
    "gpa": 3.85,
    "enrolled": True
}

print("Dictionary:", student)
print("Name:", student["name"])
print("GPA:", student.get("gpa"))
print("Keys:", list(student.keys()))
print("Values:", list(student.values()))

# Adding / Updating
student["email"] = "alice@example.com"
student["age"] = 22
print("\nUpdated dictionary:", student)

# Deleting
del student["enrolled"]
print("After deletion:", student)

# Iterating
print("\nAll key-value pairs:")
for key, value in student.items():
    print(f"  {key}: {value}")
```

## 8. Sets
Sets are unordered collections of **unique** elements. Useful for membership testing and eliminating duplicates.

```python
# Creating sets
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}

print("Set A:", set_a)
print("Set B:", set_b)

# Set operations
print("Union (A | B)       :", set_a | set_b)
print("Intersection (A & B):", set_a & set_b)
print("Difference (A - B)  :", set_a - set_b)
print("Symmetric Diff      :", set_a ^ set_b)

# Removing duplicates from a list
with_duplicates = [1, 2, 2, 3, 4, 4, 4, 5]
unique = list(set(with_duplicates))
unique.sort()
print("\nOriginal list:", with_duplicates)
print("After removing duplicates:", unique)
```

## 9. Type Conversion

```python
# Implicit and Explicit Type Conversion
x = 5        # int
y = 2.5      # float

result = x + y   # implicit conversion (int → float)
print(f"{x} + {y} = {result} (type: {type(result).__name__})")  

# Explicit conversions
print("\nExplicit Conversions:")
print(f"int('42')      = {int('42')}")
print(f"float('3.14')  = {float('3.14')}")
print(f"str(100)       = '{str(100)}'")
print(f"bool(0)        = {bool(0)}")
print(f"bool(1)        = {bool(1)}")
print(f"bool('')       = {bool('')}")
print(f"bool('hello')  = {bool('hello')}")
print(f"list((1,2,3))  = {list((1,2,3))}")
print(f"tuple([1,2,3]) = {tuple([1,2,3])}")
```

## 11. Practice Exercises

```python
# Exercise 1: Simple Calculator
def calculator(a, b, operation):
    """Perform basic arithmetic based on operator string."""
    if operation == '+':
        return a + b
    elif operation == '-':
        return a - b
    elif operation == '*':
        return a * b
    elif operation == '/':
        if b == 0:
            return "Error: Division by zero!"
        return a / b
    else:
        return "Unknown operation"

print("=== Simple Calculator ===")
for op in ['+', '-', '*', '/']:
    print(f"10 {op} 3 = {calculator(10, 3, op)}")
```

```python
# Exercise 2: Word frequency counter
text = "the quick brown fox jumps over the lazy dog the fox"
words = text.split()
frequency = {}

for word in words:
    frequency[word] = frequency.get(word, 0) + 1

print("=== Word Frequency ===")
sorted_freq = sorted(frequency.items(), key=lambda x: x[1], reverse=True)
for word, count in sorted_freq:
    print(f"  '{word}': {count}")
```

```python
# Exercise 3: Student grade summary
students = {
    "Alice": [88, 92, 79, 95, 84],
    "Bob":   [72, 68, 75, 80, 65],
    "Carol": [95, 98, 92, 97, 100],
    "Dave":  [60, 55, 70, 63, 58]
}

print("=== Student Grade Summary ===")
print(f"{'Name':<10} {'Average':>10} {'Grade':>8}")
print("-" * 30)

for name, scores in students.items():
    avg = sum(scores) / len(scores)
    if avg >= 90:
        letter = 'A'
    elif avg >= 80:
        letter = 'B'
    elif avg >= 70:
        letter = 'C'
    elif avg >= 60:
        letter = 'D'
    else:
        letter = 'F'
    print(f"{name:<10} {avg:>10.2f} {letter:>8}")
```
