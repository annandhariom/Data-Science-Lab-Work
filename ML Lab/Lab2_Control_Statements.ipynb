## 1. Conditional Statements (`if` / `elif` / `else`)
Conditional statements allow a program to choose different execution paths based on boolean expressions.

```python
# Basic if-elif-else
score = 78

if score >= 90:
    grade = 'A'
    remark = 'Excellent'
elif score >= 80:
    grade = 'B'
    remark = 'Good'
elif score >= 70:
    grade = 'C'
    remark = 'Average'
elif score >= 60:
    grade = 'D'
    remark = 'Below Average'
else:
    grade = 'F'
    remark = 'Failing'

print(f"Score: {score} → Grade: {grade} ({remark})")
```

```python
# Nested conditionals
age = 20
has_id = True

if age >= 18:
    if has_id:
        print("Access granted: Adult with valid ID")
    else:
        print("Access denied: No valid ID")
else:
    print("Access denied: Under 18")

# Ternary (inline) expression
status = "Even" if age % 2 == 0 else "Odd"
print(f"Age {age} is {status}")
```

```python
# Checking multiple conditions
temperature = 72  # Fahrenheit
humidity = 55     # percent

if 65 <= temperature <= 80 and humidity < 60:
    print("Weather: Comfortable ☀️")
elif temperature > 90 or humidity > 80:
    print("Weather: Very hot and/or humid 🥵")
elif temperature < 50:
    print("Weather: Cold 🥶")
else:
    print("Weather: Moderate")
```

## 2. `for` Loops
`for` loops iterate over any iterable: lists, tuples, strings, ranges, dictionaries.

```python
# Iterating over a list
fruits = ["apple", "banana", "cherry", "date"]
print("Fruits:")
for i, fruit in enumerate(fruits, start=1):
    print(f"  {i}. {fruit.capitalize()}")

# Iterating over a range
print("\nSquares of 1-10:")
squares = []
for n in range(1, 11):
    squares.append(n**2)
print(squares)

# Iterating over a string
print("\nLetters in 'Python':")
for char in "Python":
    print(char, end=" ")
print()
```

```python
# Iterating over a dictionary
student_scores = {"Alice": 92, "Bob": 78, "Carol": 88, "Dave": 65}

print("Student Results:")
total = 0
for name, score in student_scores.items():
    status = "PASS" if score >= 70 else "FAIL"
    print(f"  {name:<8} {score:>3}  [{status}]")
    total += score

avg = total / len(student_scores)
print(f"\nClass Average: {avg:.2f}")
```

## 3. `while` Loops
`while` loops repeat as long as a condition is `True`.

```python
# Basic while loop
count = 1
print("Counting to 5:")
while count <= 5:
    print(f"  Count: {count}")
    count += 1
print("Done!")
```

```python
# Collatz conjecture (interesting math sequence)
n = 27
steps = 0
sequence = [n]

while n != 1:
    if n % 2 == 0:
        n = n // 2
    else:
        n = 3 * n + 1
    sequence.append(n)
    steps += 1

print(f"Collatz sequence starting from 27:")
print(sequence)
print(f"Number of steps to reach 1: {steps}")
```

## 4. `break`, `continue`, and `pass`

```python
# break — exit the loop immediately
print("break example (find first even number > 5):")
numbers = [3, 7, 1, 9, 8, 4, 6]
for num in numbers:
    if num > 5 and num % 2 == 0:
        print(f"  Found: {num}")
        break
    print(f"  Checking: {num}")
```

```python
# continue — skip the current iteration
print("continue example (skip odd numbers):")
for i in range(1, 11):
    if i % 2 != 0:
        continue
    print(f"  Even: {i}")
```

```python
# pass — placeholder; does nothing
print("pass example (placeholder for future logic):")
for i in range(5):
    if i == 3:
        pass   # TODO: handle this case later
    print(f"  i = {i}")
```

## 5. Nested Loops

```python
# Multiplication table
print("Multiplication Table (1-5):")
print("    ", end="")
for j in range(1, 6):
    print(f"{j:>4}", end="")
print()
print("    " + "----" * 5)

for i in range(1, 6):
    print(f"{i} |", end="")
    for j in range(1, 6):
        print(f"{i*j:>4}", end="")
    print()
```

```python
# Pattern printing with nested loops
print("Triangle pattern:")
for i in range(1, 6):
    print("* " * i)

print("\nDiamond pattern:")
n = 5
for i in range(1, n+1):
    print(" " * (n - i) + "* " * i)
for i in range(n-1, 0, -1):
    print(" " * (n - i) + "* " * i)
```

## 6. List Comprehensions
A concise, Pythonic way to create lists from iterables with optional filtering.

```python
# Traditional loop vs list comprehension
# Traditional:
squares_loop = []
for x in range(1, 11):
    squares_loop.append(x**2)

# List comprehension:
squares_comp = [x**2 for x in range(1, 11)]

print("Squares (loop):       ", squares_loop)
print("Squares (comprehension):", squares_comp)

# With condition (filter)
evens = [x for x in range(1, 21) if x % 2 == 0]
print("\nEven numbers 1-20:", evens)

# Nested comprehension
matrix = [[i * j for j in range(1, 4)] for i in range(1, 4)]
print("\n3x3 matrix:")
for row in matrix:
    print(" ", row)

# String manipulation
words = ["hello", "world", "python", "data", "science"]
upper_long = [w.upper() for w in words if len(w) > 4]
print("\nWords longer than 4 chars (uppercase):", upper_long)
```

## 7. Functions
Functions encapsulate reusable blocks of code. Python functions support default arguments, keyword arguments, *args, and **kwargs.

```python
# Basic function
def greet(name, greeting="Hello"):
    """Return a greeting string. greeting defaults to 'Hello'."""
    return f"{greeting}, {name}!"

print(greet("Alice"))
print(greet("Bob", "Good morning"))
print(greet(greeting="Hi", name="Carol"))  # keyword arguments
```

```python
# *args and **kwargs
def summarize(*args, **kwargs):
    """Summarize positional and keyword arguments."""
    print(f"Positional args: {args}")
    print(f"Keyword args:    {kwargs}")
    if args:
        print(f"Sum of positional: {sum(args)}")

summarize(1, 2, 3, 4, name="Alice", course="DATA.110")
```

```python
# Multiple return values
def statistics(data):
    """Return mean, min, and max of a numeric list."""
    return sum(data)/len(data), min(data), max(data)

scores = [85, 92, 78, 95, 88, 73, 91]
mean, minimum, maximum = statistics(scores)
print(f"Scores: {scores}")
print(f"Mean: {mean:.2f}  |  Min: {minimum}  |  Max: {maximum}")
```

```python
# Recursive function
def factorial(n):
    """Compute n! recursively."""
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

def fibonacci(n):
    """Return nth Fibonacci number recursively."""
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

print("Factorials:")
for i in range(8):
    print(f"  {i}! = {factorial(i)}")

print("\nFirst 10 Fibonacci numbers:")
fib_seq = [fibonacci(i) for i in range(10)]
print(" ", fib_seq)
```

## 8. Lambda Functions
Small anonymous functions defined with the `lambda` keyword. Useful with `map()`, `filter()`, `sorted()`.

```python
# Lambda basics
square = lambda x: x**2
add = lambda x, y: x + y

print(f"square(7) = {square(7)}")
print(f"add(3, 4) = {add(3, 4)}")

# map() — apply a function to every element
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, numbers))
print("\nOriginal:", numbers)
print("Squared: ", squared)

# filter() — keep elements where function returns True
evens = list(filter(lambda x: x % 2 == 0, range(1, 21)))
print("Even numbers 1-20:", evens)

# sorted() with custom key
students = [("Alice", 92), ("Bob", 78), ("Carol", 88), ("Dave", 65)]
by_score = sorted(students, key=lambda s: s[1], reverse=True)
print("\nStudents ranked by score (desc):")
for rank, (name, score) in enumerate(by_score, 1):
    print(f"  #{rank} {name}: {score}")
```

## 9. Exception Handling
Python uses `try`/`except`/`else`/`finally` blocks to handle runtime errors gracefully.

```python
# Basic exception handling
def safe_divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError as e:
        return f"Error: Invalid types — {e}"
    else:
        return result
    finally:
        pass  # always runs (cleanup code goes here)

print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide(10, "two"))
```

```python
# Handling multiple exceptions, custom messages
test_inputs = ["42", "3.14", "hello", None, "100"]

print("Converting inputs to integers:")
for val in test_inputs:
    try:
        result = int(val)
        print(f"  '{val}' → {result} ✅")
    except ValueError:
        print(f"  '{val}' → ValueError: not a valid integer ❌")
    except TypeError:
        print(f"  {val}  → TypeError: cannot convert None ❌")
```

## 10. Practice Exercises

```python
# Exercise 1: FizzBuzz (classic interview question)
print("FizzBuzz (1-30):")
result = []
for n in range(1, 31):
    if n % 15 == 0:
        result.append("FizzBuzz")
    elif n % 3 == 0:
        result.append("Fizz")
    elif n % 5 == 0:
        result.append("Buzz")
    else:
        result.append(str(n))
print(", ".join(result))
```

```python
# Exercise 2: Prime number sieve
def sieve_of_eratosthenes(limit):
    """Return all prime numbers up to limit."""
    is_prime = [True] * (limit + 1)
    is_prime[0] = is_prime[1] = False
    
    for i in range(2, int(limit**0.5) + 1):
        if is_prime[i]:
            for multiple in range(i*i, limit+1, i):
                is_prime[multiple] = False
    
    return [num for num, prime in enumerate(is_prime) if prime]

primes = sieve_of_eratosthenes(100)
print(f"Prime numbers up to 100 ({len(primes)} total):")
print(primes)
```

```python
# Exercise 3: Caesar cipher
def caesar_cipher(text, shift, decode=False):
    """Encode or decode text using Caesar cipher."""
    if decode:
        shift = -shift
    result = []
    for char in text:
        if char.isalpha():
            base = ord('A') if char.isupper() else ord('a')
            shifted = (ord(char) - base + shift) % 26 + base
            result.append(chr(shifted))
        else:
            result.append(char)
    return ''.join(result)

original = "Hello, World! Python is Amazing."
shift = 13
encoded = caesar_cipher(original, shift)
decoded = caesar_cipher(encoded, shift, decode=True)

print(f"Original : {original}")
print(f"Encoded  : {encoded}")
print(f"Decoded  : {decoded}")
```
