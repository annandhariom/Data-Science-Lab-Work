## Problem 1: Confusion Matrix Metrics

```python
# Given values
TP = 50
FP = 8
TN = 32
FN = 10

# Total predictions
total = TP + FP + TN + FN

# Metrics
accuracy = (TP + TN) / total
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1_score = 2 * precision * recall / (precision + recall)

# Output
print("Total Predictions:", total)
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1_score)
```

## Problem 2: R² (Coefficient of Determination)

```python
# Given values
SSR = 420
SSE = 180

# Total Sum of Squares
SST = SSR + SSE

# R squared
R2 = SSR / SST

# Output
print("Total Sum of Squares (SST):", SST)
print("R² Value:", R2)

if R2 > 0.7:
    print("Model explains more than 70% variance")
else:
    print("Model explains less than 70% variance")
```

## Problem 3: Mean Squared Error (MSE)

```python
# Squared errors
e1 = 4
e2 = 9
e3 = 16
e4 = 25
e5 = 36

# MSE calculation
mse = (e1 + e2 + e3 + e4 + e5) / 5

# Output
print("Mean Squared Error (MSE):", mse)
```

## Problem 4: Cook’s Distance

```python
# Given values
ei2 = 9
MSE = 3
p = 3
hii = 0.2
n = 25

# Cook's Distance formula
cooks_d = ei2 / (p * MSE) * (hii / ((1 - hii) ** 2))

# Threshold
threshold = 4 / n

# Output
print("Cook's Distance:", cooks_d)
print("Threshold:", threshold)

if cooks_d > threshold:
    print("Observation is Influential")
else:
    print("Observation is NOT Influential")
```

## Problem 5: Simple Linear Regression Prediction

```python
# Given regression equation: y = mx + c
m = 2      # slope
c = 5      # intercept
x = 10     # rainfall value

# Prediction
y = m * x + c

# Output
print("Predicted Crop Yield:", y)
print("Slope shows change in output per unit input")
print("Intercept shows output when input is zero")
```
